```python
def sum_df(*args):
    dataframes = []
    for arg in args:
        csv_file = pd.read_csv(f"/kaggle/input/0db0ed38-868c-40d2-bcbd-bcedb88543cf/dataset/dataset/{arg}", index_col=0)
   #     csv_file["Dates"] = pd.to_datetime(csv_file["Dates"])
        csv_file.index.name = "Date"
        csv_file = csv_file.fillna(0)
        csv_file.index = pd.to_datetime(csv_file.index)
        dataframes.append(
            csv_file.fillna(0)
        )
    co2_mean = []
    for df in dataframes:
        co2_mean.append(float(df["CO2"].mean()))
    co2 = sum(co2_mean)/len(co2_mean)
    return reduce(lambda x, y: x.add(y, fill_value=0), dataframes), co2
```

```python
def multitable_dataframe(df, group, selected_group):
    """Return the sum of the selected tables
    """
    select = df.groupby(group)["FILE_ID"].apply(list)[selected_group]
    return sum_df(*select)
```

```python
engine_metadata = pd.read_csv("/kaggle/input/0db0ed38-868c-40d2-bcbd-bcedb88543cf/engine_metadata.csv")
```

```python
engine1, co2_1 = multitable_dataframe(engine_metadata, "ENGINE_ID", "ENGINE_1")
engine2, co2_2 = multitable_dataframe(engine_metadata, "ENGINE_ID", "ENGINE_2")
engine3, co2_3 = multitable_dataframe(engine_metadata, "ENGINE_ID", "ENGINE_3")
engine4, co2_4 = multitable_dataframe(engine_metadata, "ENGINE_ID", "ENGINE_4")
engine5, co2_5 = multitable_dataframe(engine_metadata, "ENGINE_ID", "ENGINE_5")
engine6, co2_6 = multitable_dataframe(engine_metadata, "ENGINE_ID", "ENGINE_6")
engine7, co2_7 = multitable_dataframe(engine_metadata, "ENGINE_ID", "ENGINE_7")
engine8, co2_8 = multitable_dataframe(engine_metadata, "ENGINE_ID", "ENGINE_8")
```
