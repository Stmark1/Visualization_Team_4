```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
table = pd.read_csv('../input/metadata/site_metadata.csv')

gdf = gpd.GeoDataFrame(table, geometry=gpd.points_from_xy(table.LONGITUDE, table.LATITUDE))
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

ax = world.plot(color='white', edgecolor='black')

gdf.plot(ax=ax, markersize=8, column='CUSTOMER_NAME',legend=True, cmap="Blues")
leg = ax.get_legend()
leg.set_bbox_to_anchor((1.15,1))

ax.set_axis_off()
ax.set(title='Ubicaci√≥n de las plantas por cliente')
plt.show()
```
